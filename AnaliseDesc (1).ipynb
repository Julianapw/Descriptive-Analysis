12.Identifique se há presença de valores outliers para a métrica OverallQual.
Qual método será utilizado para identificar os valores outliers? Justifique sua resposta.
## Utilizou-se o método do Intervalo Interquartílico (IQR). Ele é o padrão estatístico para definir limites técnicos de normalidade, sendo robusto contra distorções e a base matemática para a construção de Boxplots.
Quantos registros foram identificados como outliers? Essa quantidade representa qual porcentagem do total de informações?
## Foram identificados 4 registros como outliers (casas com nota 1). Essa quantidade representa aproximadamente 0,14% do total de informações.

```python
q1_over = housing['Overall Qual'].quantile(0.25)
q3_over = housing['Overall Qual'].quantile(0.75)
iqr_over = q3_over - q1_over

limite_inf_over = q1_over - 1.5 * iqr_over
limite_sup_over = q3_over + 1.5 * iqr_over


outliers_over = housing[(housing['Overall Qual'] < limite_inf_over) | (housing['Overall Qual'] > limite_sup_over)]
print(outliers_over.head())
```

13.Determine o Escore Z da variável OverallQual para 10 registros aleatórios. (Utilize o método sample()). A nota de algum desses imóveis é considerado valor outliers? Justifique sua resposta.  
## Não, para ser considerado um outlier pelo método do Escore Z, o valor deve ser maior que 3 ou menor que -3. Como a escala de OverallQual é limitada (1 a 10), as notas da amostra raramente se afastam da média o suficiente para ultrapassar esse limite estatístico.

```python
media = housing['Overall Qual'].mean()
desvio_padrao = housing['Overall Qual'].std()
amostra = housing['Overall Qual'].sample(10, random_state=42) # random_state para você poder repetir o teste
z_scores = (amostra - media) / desvio_padrao

resultado = pd.DataFrame({'Nota (Overall Qual)': amostra, 'Escore Z': z_scores})
print(resultado)
```
